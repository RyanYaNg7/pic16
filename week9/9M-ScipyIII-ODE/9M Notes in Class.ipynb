```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.integrate as spint
```

```python
def f(z, t, mu):
    dz = np.zeros(len(z))
    dz[0] = z[1]
    dz[1] = mu*(1-z[0]**2)*z[1] - z[0]
    return dz
```

```python
mu = 2
f2 = lambda z, t: f(z, t, mu)
t = np.linspace(0, 10, 100)
z = spint.odeint(f2, np.array([1,1]), t)
```

```python
plt.plot(t, z[:, 0])
```
